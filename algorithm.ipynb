# Import Library

```python
import time
import matplotlib.pyplot as plt
```

# Fibonacci Function

```python
#Using Dynamic Programming
def fib(n):
    seq = [0,1]
    if n < len(seq):
        return seq[n]
    else:
        for n in range(len(seq), n+1):
            seq.append(seq[-1]+seq[-2])
    return seq[n]

#Using Divide and Conquer
def fibDnC(n):
    return fibDnC(n-1) + fibDnC(n-2) if n > 1 else n
```

## Input n-series

```python
print('Fibonacci series?')
size = int(input())

if size < 1:
    raise ValueError('n must be a positive integer')
```

# Main Program

```python
time_dynamic = [] #To store running time of dynamic programming for each iteration
time_DnC = [] #To store running time of divide and conquer for each iteration
ns = [] #To store nsize(for x-axis plot)

for n in range(1, size+1):
    ns.append(n)

start = time.time()
for n in range(1, size+1):
    fib(n)
    finish = time.time() - start
    time_dynamic.append(finish)

mulai = time.time()
for n in range(1, size+1):
    fibDnC(n)
    selesai = time.time() - mulai
    time_DnC.append(selesai)
```

# Make a Line Plot

```python
fig = plt.subplots(1,1,figsize=(18, 9))
plt.plot(ns, time_dynamic, label='Dynamic', color='r')
plt.plot(ns, time_DnC, label='DnC', color='g', linestyle='--')
plt.title('Dynamic Programming VS DnC', fontsize=30)
plt.xlabel('n size', fontsize=20)
plt.ylabel('Runtime (s)', fontsize=20)
plt.legend()
plt.ylim([0, 15000]) #use it while the nsize > 40
#plt.savefig('grafik_fix.png')
```
